# Name Map
Create dict from tournament name to stats name to allow lookup of stats from tournament brackets. Just need to re-run if there are error names in the tournament year you're running. Update `name_map` with the correct mapping and then paste it into `simulate_tournament.py`

```python
import parse_tournament as pt
import stats_table as st
```

```python
year = 2021
t = pt.parse_tournament(year)
df = st.get_combined_stats_table(year)
```

```python
missing = []

for r in t:
    for g in r:
        t1, t2 = g[0][1], g[1][1]
        
        if t1 not in df.index:
            if t1 not in name_map:
                missing.append(t1)
            elif name_map[t1] not in df.index:
                print("Error: {}:{} map".format(t1, name_map[t1]))
            
        if t2 not in df.index:
            if t2 not in name_map:
                missing.append(t2)
            elif name_map[t2] not in df.index:
                print("Error: {}:{} map".format(t2, name_map[t2]))
            
if len(missing) == 0:
    print("All good!")
else:
    print(missing)
```
